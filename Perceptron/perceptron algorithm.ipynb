### Perceptron

<p>A perceptron is a type of artificial neuron used in machine learning and artificial intelligence, specifically in the context of neural networks.
z = w1x1 + w2x2 + b
</p>

```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
sns.set()
```

```python
data = pd.read_csv('placement.csv')
```

```python
print(data.shape)
data.head()
```

```python
sns.scatterplot(x='cgpa',y='resume_score',hue='placed',data=data)
```

```python
#creating the input and the target variable 

x = data.iloc[:,0:2]
y= data.iloc[:,-1]
```

```python
 #importing perceptron and creating its object
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(x,y)
```

```python
# this gives the weights of the input
p.coef_
```

```python
# this gives the bias
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.values,y.values,clf=p,legend=2)
```
